```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# create a model class that inherits nn.modules

class Model (nn.Module):
  # input layer (4 features of the flower)--> Hidden layer 1 ( number of neurones) --> Hidden layer2 --> output layer (the class of the iris)
  def __init__(self, in_features=4, h1=8, h2 = 9, out_feature=3):
    super().__init__()# instantiate our nn.Module
    self.fc1 = nn.Linear(in_features, h1)
    self.fc2 = nn.Linear(h1, h2)
    self.out = nn.Linear(h2, out_feature)

  def forward(self, x):
    x = F.relu(self.fc1(x))
    x = F.relu(self.fc2(x))
    x = self.out(x)
    return x
```

```python
# pick a manual seed for randomization
torch.manual_seed(41)
# create an instance of the model
model=Model()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
url='https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv'
my_df=pd.read_csv(url)
my_df.head()
```

```python
my_df['species'] = my_df['species'].replace('setosa',0.0)
my_df['species'] = my_df['species'].replace('versicolor',1.0)
my_df['species'] = my_df['species'].replace('virginica',2.0)
my_df
```

```python
# train test and split
x=my_df.drop('species', axis=1)
y=my_df['species']
```

```python
x=x.values
y=y.values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=41)
```

```python
# convert x features into a tensor
x_train=torch.FloatTensor(x_train)
x_test=torch.FloatTensor(x_test)
```

```python
# convert y labels to tensor long
y_train=torch.LongTensor(y_train)
y_test=torch.LongTensor(y_test)
```

```python
# set the criterion to our model to mesure the error
criterion = nn.CrossEntropyLoss()
# choose Adam optimizer, learning rate (if our error doesnt go down after a bunch of iterations (epochs))
optimizer=torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
#train our model
# epochs? one run through all the training data in our network
epochs = 100
# the loss must be decreasing each epoch
losses=[]

for i in range(epochs):
 # get prediction
 y_pred= model.forward(x_train)
 # mesure the loss
 loss=criterion(y_pred, y_train) # predicted values vs the y_train
 # keep track of the loss
 losses.append(loss.detach().numpy())


 # print every 10 epochs
 if i % 10 == 0 :
  print (f'Epoch : {i} and loss : {loss}')

 # back propagation: take the error rate of foward propagation and feed it back through the network to fine tune the weights
 optimizer.zero_grad()
 loss.backward()
 optimizer.step()
```

```python
# garph the loss

plt.plot (range(epochs),losses)
plt.ylabel("loss")
plt.xlabel("epoch")
```
